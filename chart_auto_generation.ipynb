```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from dateutil.parser import parse
import datetime
%matplotlib inline
```

```python
df=pd.read_csv('apr11.csv',parse_dates=['dateRep'])
pd.set_option('display.max_rows',None)
df.head()
```

```python
df['date']=[d.strftime('%x') for d in df['dateRep']]
df.head()
```

```python
def df_filtered(country_code):
    country_filter=df['geoId']==country_code
    df2=df[country_filter]
    df2.sort_index(ascending=False,inplace=True)
    df2.reset_index(inplace=True,drop=True)
    return df2
```

```python
us=df_filtered('US')
us
```

```python
def bar_plot(country_code,left):
    country_filter=df['geoId']==country_code
    df2=df[country_filter]
    df2.sort_index(ascending=False,inplace=True)
    df2.reset_index(inplace=True)
    
    x=df2.index
    y1=df2.cases
    y2=df2.deaths

    width=0.3
    plt.style.use('fivethirtyeight')
    plt.figure(figsize=(20,5))
    plt.bar(x-width/2,y1,color='#444444',width=width,label='Cases')
    plt.bar(x+width/2,y2,color='#008fd5',width=width,label='Deaths')
    plt.xticks(ticks=df2.index,
           labels=df2['date'],
           rotation =90)
    plt.xlim(left=left)
    plt.tight_layout()
    plt.legend()
    plt.title(f'Number of cases and deaths by COVID-19 in {country_code}')
```

```python
bar_plot('US',70)
```

```python
bar_plot('DE',0)
```
